# ISIN ( ) Function.   Most Destructive Death Over in IPL (S/R)

```python
# Highest Death Over in 16-20 Overs ( Min 200 Balls i.e More Than 20 Matches)
```

```python
matches.head(5)
```

```python
death_overs = matches[matches['over'] > 15]
death_overs
```

```python
batsman_list = death_overs.groupby("batsman")["batsman_runs"].count()
x = batsman_list > 200 
all_batsman = batsman_list[x].index.tolist()
```

```python
all_batsman
```

```python
runs_list = death_overs.groupby("batsman")["batsman_runs"].sum()
y = runs_list > 200 
runs_list[y]
```

```python
Final = death_overs[death_overs["batsman"].isin(all_batsman)]
```

```python
Runs = Final.groupby("batsman")["batsman_runs"].sum()
Balls = Final.groupby("batsman")["batsman_runs"].count()
```

```python
SR = Runs/Balls * 100
SR
```
